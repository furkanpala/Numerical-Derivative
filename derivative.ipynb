```python
import numpy as np
import matplotlib.pyplot as plt

def derivative(f,order,h,x0):
    if order == 2:
       return (f(x0 + h) - f(x0 - h)) / (2*h) 
    elif order == 4:
        return (1/(12*h))*(f(x0-2*h) - (8*f(x0-h)) + (8*f(x0+h)) - f(x0 + 2*h))  
    elif order == 6:
        return (1/(4*h))*(((-1/15)*f(x0 - 3 * h)) + ((3 / 5) * f(x0 - 2 * h)) - 3 * f(x0 - h) )
    else:
        print("Invalid order")
        
f = lambda x : np.exp(x) * np.cos(2 * x)
df = lambda x : np.exp(x) * (np.cos(2 * x) - 2 * np.sin(2 * x))

h = np.array([np.float_power(10, -(i+1)) for i in range(10)]);
result = derivative(f,2,h,1)
error = np.abs(df(1) - result)



for i in range(10):
    print("Result: %.10f \t Error: %.10f \t h: %e" %(result[i],error[i],h[i]))


plt.plot(h,error,"r",label="second order")


result = derivative(f,4,h,1)
error = np.abs(np.subtract(df(1), result))

plt.plot(h,error,"b", label = "fourth order")



plt.xscale("log")
plt.yscale("log")

plt.grid(True)
plt.legend()

plt.savefig("fig1")
```
